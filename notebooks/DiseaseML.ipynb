```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
df_train = pd.read_csv('training.csv')
df_test = pd.read_csv('Testing.csv')
```

```python
df_train
```

```python
df_train = df_train.drop('Unnamed: 133', axis = 1)
```

```python
df_train.prognosis.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X, y = df_train.drop('prognosis', axis=1), df_train['prognosis']

yt = df_test.drop(["prognosis"],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4, random_state=9)
```

#### Training 6 Models

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier()

forest.fit(X_train, y_train)
```

```python
from sklearn.naive_bayes import GaussianNB

nb_clf = GaussianNB()

nb_clf.fit(X_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = GradientBoostingClassifier()

gb_clf.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train_scaled, y_train)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train_scaled, y_train)
```

#### Accuracy

```python
forest.score(X_test, y_test)
```

```python
nb_clf.score(X_test, y_test)
```

```python
gb_clf.score(X_test,y_test)
```

```python
knn.score(X_test_scaled, y_test)
```

```python
log.score(X_test_scaled, y_test)
```

```python
svc.score(X_test_scaled, y_test)
```

#### Recall Value

```python
from sklearn.metrics import recall_score

y_preds = forest.predict(X_test)
print('Forest:', recall_score(y_test, y_preds, average='macro'))

y_preds = nb_clf.predict(X_test)
print('NB:', recall_score(y_test, y_preds, average='macro'))

y_preds = gb_clf.predict(X_test)
print('GB:', recall_score(y_test, y_preds, average='macro'))

y_preds = knn.predict(X_test_scaled)
print('KNN:', recall_score(y_test, y_preds, average='macro'))

y_preds = log.predict(X_test_scaled)
print('Log:', recall_score(y_test, y_preds, average='macro'))

y_preds = svc.predict(X_test_scaled)
print('SVC:', recall_score(y_test, y_preds, average='macro'))
```
